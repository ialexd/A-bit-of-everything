```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns;sns.set()
```

```python
def w0():
    x = np.linspace(-3,3,1000)
    return np.power(np.pi,-1/4)* np.exp(-x*x/2)

def w1():
    x = np.linspace(-3,3,1000)
    return np.power(np.pi,-1/4)* np.power(2,1/2)*x*np.exp(-x*x/2)

def En(n):
    return n+(1/2)
```

```python
osc = {0:w0() , 1:w1()}
```

```python
def wave(n):
    if n in osc:
        return osc[n]
    num = np.power(2/n,1/2)*wave(n-1)-np.power((n-1)/n,1/2)*wave(n-2)
    osc[n] = num
    return num
```

```python
w = [wave(i) for i in range(3)]
```

```python
for wave in w:
    plt.plot(np.linspace(-3,3,1000),wave);
```

```python
wq = pd.DataFrame(np.transpose(w))
```
